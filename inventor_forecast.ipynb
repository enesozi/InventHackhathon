```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import scipy.stats
import datetime as dt
import sklearn
import statsmodels
%matplotlib inline
```

```python
#get data from file
data = pd.read_csv('InventoryPosition.csv',delimiter=';')
```

```python
#compute product price
data['ProductPrice'] = (data['SalesRevenue']/data['SalesQuantity']).round(2)
#group data according to columns
grouped = data.groupby(["ProductCode","Date","StoreCode", "ProductPrice","SalesQuantity"])
#discard the rows containig negative values
sqDiscarded = grouped.filter(lambda x: x["SalesQuantity"] > 0) 
sqDiscardedCopy = sqDiscarded
```

```python
#gets the data and returns mean, minimum value, maximum value within
#a confidence interval 0.95
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0*np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * sp.stats.t._ppf((1+confidence)/2., n-1)
    return m, m-h, m+h
```

```python
#group data according to product codes
sqDiscardedCopy = sqDiscardedCopy.groupby(["ProductCode"])
sqDiscardedCopy.size()
```

```python
group_price_low = {}
group_price_high = {}
#iterate over groups
#compute the maximum and minimum values for each product type
for name, group in sqDiscardedCopy :
    x = group[['ProductPrice']].as_matrix()
    a = mean_confidence_interval(x)
    group_price_low[name] = a[1][0]
    group_price_high[name] = a[2][0]
```

```python
#cast the gruped data to data frame
ss = sqDiscarded.filter(lambda x: True)
```

```python
#get product types from input file
dataprods  = pd.read_csv('Products.csv',)
#convert dataframe to array
sss = np.array(ss)
ds = np.array(dataprods)
#iterate and discard the rows containing price out of the confidenece interval
l = len(ds[:,:1])
for i in range(l):
    prodName = ds[:,:1][i][0]
    ind = np.where(np.logical_and(sss[:,2] == prodName ,(sss[:,5] > group_price_low[prodName]),(sss[:,5] < group_price_high[prodName])))
    sss = np.delete(sss, ind, axis=0)
sss    
```
